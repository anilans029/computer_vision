# **Drawing shapes on the images**

image.png

```python
rectangle_draw = cv2.rectangle(car_img, (50,10),(120,100),(0,255,0),3)
                                        ### (top-left, bottom-right, color, thickness)
cv2_imshow(rectangle_draw)
```

## **To write on images**

image.png

image.png

```python
font = cv2.FONT_ITALIC
cv2.putText(rectangle_draw, "Rectangle",(80,40),font, 0.5, (255,0,0),2)
cv2_imshow(rectangle_draw)
```

image.png

```python
circle_img = car_img
cv2.circle(circle_img,(70,63), 40, (0,0,255), -1)
cv2_imshow(circle_img)
```
